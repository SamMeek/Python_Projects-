```python
import random

#Variables
text = "Difficulty"
max_num = 0
attempts = 0

#Menu
print("Welcome to the Number Guessing Game!")
print(text.center(20, '*'))
print("1. Easy")
print("2. Medium")
print("3. Hard")
print("4. Extreme")
choice = int(input("Your Choice = "))

if choice == 1:
  max_num = 25
  attempts = 5
elif choice == 2:
  max_num = 50
  attempts = 10
elif choice == 3:
  max_num = 100
  attempts = 15
else:
  max_num = 250
  attempts = 20

number = random.randint(1, max_num)

while attempts > 0:
  guess = int(input(f"Guess a number between 1 and {max_num}: "))
  attempts -= 1
  print(f"{attempts} Attempts left!")

  if guess < number:
    print("Too low! Try again.")
  elif guess > number:
    print("Too high! Try again.")
  else:
    score = attempts * 2
    print(f"Congratulations! You guessed the Correct number {number}, Score =  {score}.")
    break
```
