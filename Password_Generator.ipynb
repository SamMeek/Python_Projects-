# **Password Generator**

```python
import random
print ("******MENU******")
print ("1. 8 Bits")
print ("2. 12 Bits")
print ("3. 16 Bits")
print ("****************")
choice = int(input("Enter your choice: "))
match choice:
  case 1:
    print ("8 Bits Strong Password Generator")
    chrs = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*()_+"
    password_list = ""
    for char in range(0, 8):
      password_list = random.choice(chrs) + password_list

    print (f"Your 8-Bits Strong Password is : {password_list}")
  case 2:
    print ("12 Bits Strong Password Generator")
    chrs = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*()_+"
    password_list = ""
    for char in range(0, 12):
      password_list = random.choice(chrs) + password_list

    print (f"Your 12-Bits Strong Password is : {password_list}")
  case 3:
    print ("16 Bits Strong Password Generator")
    chrs = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*()_+"
    password_list = ""
    for char in range(0, 16):
      password_list = random.choice(chrs) + password_list

    print (f"Your 16-Bits Strong Password is : {password_list}")
  case _:
    print ("Invalid Choice")
```
